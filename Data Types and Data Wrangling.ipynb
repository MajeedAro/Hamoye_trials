### Split and Merge

```python
fuel_data1= data.iloc[0:19000].reset_index(drop=True)
fuel_data2= data.iloc[19000:].reset_index(drop=True)
```

```python
assert len(data)==(len(fuel_data1)+len(fuel_data2))
```

Type of Merge: pd.merge(fuel_data1, fuel_data2, **how=?**


**Inner**: Removes row not in both Dataframes

**Outer**: Add all rows in both Dataframes

**Left**: Removes all rows in the right that doesnt match with the left and keep all rows in the left 

**Right**: Removes all rows in the left that doesnt match with the left and keep all rows in the right 

```python
pd.merge(fuel_data1, fuel_data2, how='outer')
```
