```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import random
```

```python
fuel_data=pd.read_csv('Dataset.csv')

fuel_data.isna().sum()
```

```python
plt.figure(figsize=(8,5))
plt.xticks(rotation=90)

fuel_unit=pd.DataFrame({'unit':['BEL','GAL','GRAMSU','KGU','MCF','MMBTU','MWDTH','MWHTH','TON'],
                        'count': [7998,84,464,110,11354,180,95,100,8958]})


sns.barplot(data=fuel_unit, x='unit', y='count')
plt.xlabel("Fuel Unit")
```

```python
g=sns.barplot(data=fuel_unit, x='unit', y='count')

g.set_yscale('log')
g.set_ylim(1,12000)
plt.xlabel('Fuel Data')
```

```python
sample_df=fuel_data.sample(n=50,random_state =4)

sns.regplot(x=sample_df['utility_id_ferc1'], y=sample_df['fuel_cost_per_mmbtu'], fit_reg=False)
```

```python
#Box Plot

sns.boxplot(x='fuel_type_code_pudl', y='utility_id_ferc1', data=fuel_data, palette=['m','c'])
```

```python
#KDE Plot

sns.kdeplot(sample_df['fuel_cost_per_unit_burned'], shade=True, color='r')
```
